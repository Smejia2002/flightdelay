# Documentacion y Demo del MVP

Objetivo: explicar claramente que hace y que NO hace el MVP.


## 1. Flujo general
Datos -> Modelo -> API
- Dataset original: vuelos + clima
- Limpieza y dataset model-ready
- Feature engineering con variables reconstruibles
- Modelo Logistic Regression
- API /predict con enriquecimiento de clima


## 2. Guion de demo (60-120s)
1) Presentar el problema: predecir retraso >=15 min.
2) Mostrar la entrada del API (JSON).
3) Explicar el enriquecimiento con clima (API externa).
4) Ejecutar /predict y mostrar salida.
5) Cerrar con limitaciones y mejoras futuras.


## 3. Ejemplo real /predict
(Se usa el ejemplo del contrato; la URL real se define en la fase 06).
```json
{
  "aerolinea": "AZ",
  "origen": "GIG",
  "destino": "GRU",
  "fecha_partida": "2025-11-10T14:30:00",
  "distancia_km": 350
}
```
Respuesta esperada:
```json
{
  "prevision": "Puntual",
  "probabilidad": 0.65
}
```


## 4. Limitaciones del MVP
- No incluye optimizacion avanzada ni tuning de hiperparametros.
- Usa datos historicos; cambios operativos pueden reducir el rendimiento.
- Depende de un servicio externo de clima.
- Logistic Regression con convergencia limitada (max_iter=200).


## 5. Mejoras futuras
- Ajustar hiperparametros y probar modelos mas robustos.
- Calibrar probabilidades y definir umbrales de negocio.
- Monitoreo de drift y retraining continuo.
- Cache de clima y tolerancia a fallos del proveedor.
